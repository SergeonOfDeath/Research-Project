```python
# Define nodes and edges
nodes = ['Reason', 'Domestic Violence', 'Infidelity', 'Lack of Commitment', 'Conflict', 'Unprepared Marriage', 'Financial Issue', 'Substance Abuse', 'Physical Incapacity', 'Divorce', 'Neutral', 'Negative Impact', 'Lives Improved', 'Socially', 'Emotionally', 'Financially', 'Physically']
edges = [
    ('Reason', 'Domestic Violence'),
    ('Reason', 'Infidelity'),
    ('Reason', 'Lack of Commitment'),
    ('Reason', 'Conflict'),
    ('Reason', 'Unprepared Marriage'),
    ('Reason', 'Financial Issue'),
    ('Reason', 'Substance Abuse'),  
    ('Reason', 'Physical Incapacity'), 
    ('Domestic Violence', 'Divorce'),
    ('Infidelity', 'Divorce'),
    ('Domestic Violence', 'Divorce'),
    ('Infidelity', 'Divorce'),
    ('Lack of Commitment', 'Divorce'),
    ('Conflict', 'Divorce'),
    ('Unprepared Marriage', 'Divorce'),
    ('Financial Issue', 'Divorce'),
    ('Substance Abuse', 'Divorce'),  
    ('Physical Incapacity', 'Divorce'), 
    ('Divorce', 'Neutral'),
    ('Divorce', 'Negative Impact'),
    ('Divorce', 'Lives Improved'),
    ('Negative Impact', 'Socially'),
    ('Negative Impact', 'Emotionally'),
    ('Negative Impact', 'Financially'),
    ('Negative Impact', 'Physically'),
    ('Domestic Violence','Physically'),
    ('Infidelity','Socially'),
    ('Financial Issue', 'Financially'),
    ('Physical Incapacity', 'Physically'),
    ('Conflict', 'Socially'),
    ('Socially','Physically'),
    ('Socially','Financially'),
    ('Socially','Emotionally'),
    ('Financially','Emotionally'),
    ('Financially','Physically'),
    ('Emotionally','Physically')
]

# Create the graph
G = nx.DiGraph()  # Use a directed graph to show causal relationships
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Define layout and visualize the network
pos = nx.spring_layout(G)
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=4000, node_color='lightblue', font_size =8,cmap='coolwarm')  # Customize as needed
plt.title("Network Analysis of Divorce Causes, Consequences, and Impacts")
plt.show()

```

```python
# Degree Centrality
degree_centrality = nx.degree_centrality(G)
print("\nDegree Centrality:")
for node, centrality in degree_centrality.items():
    print(f"{node}: {centrality:.4f}")

# Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)
print("\nBetweenness Centrality:")
for node, centrality in betweenness_centrality.items():
    print(f"{node}: {centrality:.4f}")

# Closeness Centrality
closeness_centrality = nx.closeness_centrality(G)
print("\nCloseness Centrality:")
for node, centrality in closeness_centrality.items():
    print(f"{node}: {centrality:.4f}")

# Clustering Coefficient
clustering_coefficient = nx.average_clustering(G)
print(f"\nAverage Clustering Coefficient: {clustering_coefficient:.4f}")

# PageRank
pagerank = nx.pagerank(G)
print("\nPageRank:")
for node, rank in pagerank.items():
    print(f"{node}: {rank:.4f}")
```

```python
# Perform community detection using the Louvain method
communities = list(greedy_modularity_communities(G))

# Create a new graph with communities as nodes
community_graph = nx.Graph()
for i, community in enumerate(communities):
    community_graph.add_node(f'Community {i + 1}', nodes=community)
    for node in community:
        community_graph.add_edge(f'Community {i + 1}', node)

# Draw the community graph
pos = nx.spring_layout(community_graph, seed=42)
plt.figure(figsize=(12, 8))
nx.draw(community_graph, pos, with_labels=True, font_weight='bold', node_size=800, node_color='skyblue', font_size=8, edge_color='gray', width=1.5)

# Show the plot
plt.show()

# Print community sizes
for i, community in enumerate(communities):
    print(f"Community {i + 1} Size: {len(community)}")
    print(f"Nodes: {community}\n")

modularity = nx.community.modularity(G, communities)
print(f"Modularity: {modularity:.4f}")

# Visualization of each community
for i, community in enumerate(communities):
    subgraph = G.subgraph(community)
    pos = nx.spring_layout(subgraph, seed=42)
    plt.figure(figsize=(8, 6))
    nx.draw(subgraph, pos, with_labels=True, font_weight='bold', node_size=600, node_color='skyblue', font_size=8, edge_color='gray', width=1.5)
    plt.title(f"Community {i + 1}")
    plt.show()

# Identify bridge nodes
bridge_nodes = set()
for edge in G.edges():
    for community in communities:
        if edge[0] in community and edge[1] not in community:
            bridge_nodes.add(edge[0])
        elif edge[1] in community and edge[0] not in community:
            bridge_nodes.add(edge[1])

print(f"Bridge Nodes: {bridge_nodes}")

# Example: Calculate average degree centrality for each community
for i, community in enumerate(communities):
    avg_degree_centrality = sum(nx.degree_centrality(G).get(node, 0) for node in community) / len(community)
    print(f"Community {i + 1} Average Degree Centrality: {avg_degree_centrality:.4f}")
```

**Community Detection**

**1.Community Sizes:**\
-Community 1 has a larger size (12 nodes) compared to Community 2 (5 nodes). This suggests that there is a significant diversity in the types of nodes within the larger community.

**2.Community Connectivity:**\
-The presence of bridge nodes, including "Negative Impact," "Divorce," and others, indicates that certain nodes play a crucial role in connecting the two communities. These bridge nodes could be influencing factors that are relevant across various aspects of divorce.

**3.Modularity:**\
-The modularity value of 0.3114 indicates a moderate level of community structure within the network. It suggests that the division into communities is meaningful, but there may still be some degree of overlap or interconnectivity between the communities.

**4.Average Degree Centrality:**\
-The average degree centrality for Community 1 is 0.2188, while for Community 2, it is higher at 0.3250. This suggests that, on average, nodes in Community 2 have more connections, indicating a potentially more interconnected structure within that community.

**Conclusion:**

-The network exhibits a meaningful community structure, with two distinct communities identified.\
-Community 1 is larger and includes nodes related to reasons for divorce, consequences, and impacts. This community is more diverse and covers a broader range of aspects related to divorce.\
-Community 2 is smaller and includes nodes related to negative impacts, social aspects, emotions, finances, and physical well-being. This community seems to focus more specifically on the consequences and impacts of divorce.\
-Bridge nodes like "Negative Impact," "Divorce," and others connect the two communities, suggesting shared factors or relationships that cut across different aspects of divorce.\
-The average degree centrality indicates that nodes in Community 2 are, on average, more connected than those in Community 1.

These findings provide a valuable glimpse into the internal structure of the network, offering insights into how different aspects of divorce are interconnected and how specific nodes serve as bridges between various elements.

```python
# Hubs and Authorities analysis
hubs, authorities = nx.hits(G, max_iter=100, normalized=True)

# Print the results
print("\nHubs:")
for node, hub_score in hubs.items():
    print(f"{node}: {hub_score:.4f}")

print("\nAuthorities:")
for node, auth_score in authorities.items():
    print(f"{node}: {auth_score:.4f}")

# Identify top hubs and authorities
top_hubs = sorted(hubs.items(), key=lambda x: x[1], reverse=True)[:5]
top_authorities = sorted(authorities.items(), key=lambda x: x[1], reverse=True)[:5]

print("\nTop Hubs:")
for node, hub_score in top_hubs:
    print(f"{node}: {hub_score:.4f}")

print("\nTop Authorities:")
for node, auth_score in top_authorities:
    print(f"{node}: {auth_score:.4f}")



# Visualization of hubs and authorities
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)  #can use a different layout
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=800, font_size=8, edge_color='gray', width=1.5)

# Extract nodes from the tuples in the top_hubs and top_authorities lists
top_hubs_nodes = [node for node, _ in top_hubs]
top_authorities_nodes = [node for node, _ in top_authorities]

nx.draw_networkx_nodes(G, pos, nodelist=top_hubs_nodes, node_color='red', node_size=800, label='Hubs')
nx.draw_networkx_nodes(G, pos, nodelist=top_authorities_nodes, node_color='blue', node_size=800, label='Authorities')
plt.legend()
plt.title("Network with Top Hubs and Authorities Highlighted")
plt.show()

# Example: Correlation with Degree Centrality
degree_centrality = nx.degree_centrality(G)

# Extract hub and authority scores as lists
hub_scores = [hubs[node] for node in G.nodes()]
authority_scores = [authorities[node] for node in G.nodes()]

# Remove nodes with zero hub or authority scores
filtered_nodes = [node for node in G.nodes() if hubs[node] != 0 and authorities[node] != 0]
filtered_hub_scores = [hubs[node] for node in filtered_nodes]
filtered_authority_scores = [authorities[node] for node in filtered_nodes]

# Compute correlation
hub_correlation = np.corrcoef(filtered_hub_scores, [degree_centrality[node] for node in filtered_nodes])[0, 1]
authority_correlation = np.corrcoef(filtered_authority_scores, [degree_centrality[node] for node in filtered_nodes])[0, 1]

print(f"\nCorrelation with Degree Centrality:")
print(f"Hub Score: {hub_correlation:.4f}")
print(f"Authority Score: {authority_correlation:.4f}")
```

**2.Top Authorities:**\

```python
# Perform community detection using the Louvain method
communities = list(greedy_modularity_communities(G))

# Create a new graph with communities as nodes
community_graph = nx.Graph()
for i, community in enumerate(communities):
    community_graph.add_node(f'Community {i + 1}', nodes=community)
    for node in community:
        community_graph.add_edge(f'Community {i + 1}', node)

# Draw the community graph
pos = nx.spring_layout(community_graph, seed=85)
plt.figure(figsize=(13,11))
nx.draw(community_graph, pos, with_labels=True, font_weight='bold', node_size=3000, node_color='skyblue', font_size=9, edge_color='gray', width=1.5)

# Show the plot
plt.show()

# Print community sizes
for i, community in enumerate(communities):
    print(f"Community {i + 1} Size: {len(community)}")
    print(f"Nodes: {community}\n")

modularity = nx.community.modularity(G, communities)
print(f"Modularity: {modularity:.4f}")

# Visualization of each community
for i, community in enumerate(communities):
    subgraph = G.subgraph(community)
    pos = nx.spring_layout(subgraph, seed=42)
    plt.figure(figsize=(8, 6))
    nx.draw(subgraph, pos, with_labels=True, font_weight='bold', node_size=3500, node_color='skyblue', font_size=9, edge_color='gray', width=1.5)
    plt.title(f"Community {i + 1}")
    plt.show()

# Identify bridge nodes
bridge_nodes = set()
for edge in G.edges():
    for community in communities:
        if edge[0] in community and edge[1] not in community:
            bridge_nodes.add(edge[0])
        elif edge[1] in community and edge[0] not in community:
            bridge_nodes.add(edge[1])

print(f"Bridge Nodes: {bridge_nodes}")

# Example: Calculate average degree centrality for each community
for i, community in enumerate(communities):
    avg_degree_centrality = sum(nx.degree_centrality(G).get(node, 0) for node in community) / len(community)
    print(f"Community {i + 1} Average Degree Centrality: {avg_degree_centrality:.4f}")
```
